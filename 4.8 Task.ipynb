```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Importing Data orders_products_combined
df_orders_products_derived = pd.read_pickle(r'C:\Users\steph\OneDrive\Documents\May 2022 Instacart Basket Analysis\02 Data\Prepared Data\orders_products_derived.pkl')
```

```python
# 2 Finding the aggregated mean
df_orders_products_derived.groupby('department_id').agg({'order_number': ['mean']})
```

# 3 The results differ a bit from the sub set, there are means that are close and there are some that have a bigger difference. 

```python
# Aggregating data with transform and creating a max_order column
df_orders_products_derived['max_order'] = df_orders_products_derived.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking new column
df_orders_products_derived.head(15)
```

```python
# 4 Creating customer loyalty flag descriptions
df_orders_products_derived.loc[df_orders_products_derived['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_orders_products_derived.loc[(df_orders_products_derived['max_order'] <= 40) & (df_orders_products_derived['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_orders_products_derived.loc[df_orders_products_derived['max_order'] <= 10,  'loyalty_flag'] = 'New customer'
```

```python
# Checking accuracy of loyalty flags
df_orders_products_derived['loyalty_flag'].value_counts(dropna = False)
```

```python
df_orders_products_derived[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# 5 Statistics of product prices for by customer category
df_orders_products_derived.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Loyal customers spend less than both new and regular customers. New customers spend the most.

```python
# 6 Aggregating data with transform and creating a Spending column
df_orders_products_derived['Spending'] = df_orders_products_derived.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking new column
df_orders_products_derived.head(10)
```

```python
df_orders_products_derived['Spending'].value_counts(dropna = False)
```

```python
df_orders_products_derived.groupby('Spending')
```

```python
# 6 Creating Spending flag descriptions and column
df_orders_products_derived.loc[df_orders_products_derived['Spending'] > 10, 'Spending_flag'] = 'High spender'
```

```python
df_orders_products_derived.loc[df_orders_products_derived['Spending'] <= 10, 'Spending_flag'] = 'Low spender'
```

```python
# Checking accuracy of Spending flags
df_orders_products_derived['Spending_flag'].value_counts(dropna = False)
```

```python
df_orders_products_derived[['user_id', 'Spending', 'Spending_flag']].head(60)
```

```python
# 6 Statistics of spending for by spending category
df_orders_products_derived.groupby('Spending_flag').agg({'Spending': ['mean', 'min', 'max']})
```

```python
# 7 Finding the aggregated mean of days_since_
df_orders_products_derived.groupby('user_id').agg({'days_since_prior_order': ['median']})
```

```python
# Aggregating data with transform and creating a Order Frequency column
df_orders_products_derived['Order_frequency'] = df_orders_products_derived.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking new column
df_orders_products_derived.head(10)
```

```python
# 4 Creating customer loyalty flag descriptions
df_orders_products_derived.loc[df_orders_products_derived['Order_frequency'] > 20, 'Frequency_flag'] = 'Non-frequent customer'
```

```python
df_orders_products_derived.loc[(df_orders_products_derived['Order_frequency'] > 10) & (df_orders_products_derived['Order_frequency'] <= 20), 'Frequency_flag'] = 'Regular customer'
```

```python
df_orders_products_derived.loc[df_orders_products_derived['Order_frequency'] <= 10,  'Frequency_flag'] = 'Frequent customer'
```

```python
# Checking accuracy of loyalty flags
df_orders_products_derived['Frequency_flag'].value_counts(dropna = False)
```
